## Unsupervised Learning (K-Means Clustering)

### Subtask:
Apply K-Means clustering for customer segmentation. First, use the Elbow method to determine the optimal number of clusters. Then, perform clustering and add the generated cluster labels as a new feature to the dataset. Visualize the distribution of these clusters.


**Reasoning**:
I will import the necessary libraries, implement the Elbow method to determine the optimal number of clusters by fitting KMeans models and collecting inertia values, and then plot these values to visualize the elbow point. This addresses the first few instructions of the subtask.



```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Initialize an empty list to store inertia values
inertia_values = []

# Loop through a range of possible cluster numbers (e.g., from 1 to 10)
k_range = range(1, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10) # n_init is explicitly set to avoid future warning
    kmeans.fit(X_processed) # Fit to the preprocessed data
    inertia_values.append(kmeans.inertia_)

# Plot the inertia values against the number of clusters (Elbow Method)
plt.figure(figsize=(10, 6))
plt.plot(k_range, inertia_values, marker='o')
plt.title('Elbow Method for Optimal K')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia')
plt.xticks(k_range)
plt.grid(True)
plt.show()

print("Elbow method plot generated. Please observe the plot to determine the optimal number of clusters.")
```

**Reasoning**:
Based on the elbow plot from the previous step, which shows the inertia significantly decreasing and then leveling off around 3 or 4 clusters, I will choose 3 as the optimal number of clusters. I will then perform K-Means clustering with this number, assign the cluster labels to the original DataFrame, and visualize the distribution of these new clusters.



```python
optimal_k = 3 # Chosen based on observing the Elbow plot where the decrease in inertia starts to slow down significantly

# 8. Initialize a KMeans model with the chosen optimal number of clusters
kmeans_model = KMeans(n_clusters=optimal_k, random_state=42, n_init=10)

# 9. Fit the KMeans model to X_processed and predict the cluster labels
cluster_labels = kmeans_model.fit_predict(X_processed)

# 10. Add the generated cluster labels as a new column named 'Cluster' to the original df DataFrame
df['Cluster'] = cluster_labels

print(f"First 5 rows of df with new 'Cluster' column:\n{df.head()}")
print(f"\nCluster counts:\n{df['Cluster'].value_counts()}")

# 11. Visualize the distribution of the created clusters using a count plot
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Cluster', hue='Cluster', palette='viridis', legend=False)
plt.title('Distribution of Customer Clusters (K-Means)')
plt.xlabel('Cluster')
plt.ylabel('Count')
plt.show()

print("K-Means clustering performed and cluster distribution visualized successfully.")
```

## Supervised Learning (Churn Prediction)

### Subtask:
Train and evaluate multiple supervised learning models for churn prediction: Random Forest, Decision Tree, and Logistic Regression. Compare their performance using metrics such as Accuracy, Precision, Recall, F1 Score, and ROC AUC. Generate confusion matrices and ROC curves for visualization.


**Reasoning**:
I will import the required libraries, initialize a dictionary for models and metrics, then iterate through Random Forest, Decision Tree, and Logistic Regression models. For each model, I will train it, make predictions, calculate and print evaluation metrics, display a confusion matrix, and plot the ROC curve as per the instructions.



```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns

# 2. Create a dictionary to store the trained models and another dictionary to store their performance metrics.
models = {}
metrics = {}

# 3. Instantiate each model with random_state=42 for reproducibility.
rf_model = RandomForestClassifier(random_state=42)
dt_model = DecisionTreeClassifier(random_state=42)
lr_model = LogisticRegression(random_state=42, solver='liblinear') # 'liblinear' is a good choice for smaller datasets and handles L1/L2 penalties

model_list = {
    "Random Forest": rf_model,
    "Decision Tree": dt_model,
    "Logistic Regression": lr_model
}

for name, model in model_list.items():
    print(f"\n--- Training and Evaluating {name} ---")

    # a. Train the model
    model.fit(X_train, y_train)

    # b. Make predictions on X_test
    y_pred = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test)[:, 1] # Probability of the positive class

    # c. Calculate and print evaluation metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    f1 = f1_score(y_test, y_pred)
    roc_auc = roc_auc_score(y_test, y_pred_proba)

    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1:.4f}")
    print(f"ROC AUC Score: {roc_auc:.4f}")

    # d. Store these metrics in the metrics dictionary.
    metrics[name] = {
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1 Score': f1,
        'ROC AUC': roc_auc
    }
    models[name] = model

    # e. Generate and display a confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['No Churn', 'Churn'], yticklabels=['No Churn', 'Churn'])
    plt.title(f'Confusion Matrix for {name}')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.show()

    # f. Calculate FPR, TPR and plot ROC curve
    fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
    roc_auc_val = auc(fpr, tpr)

    plt.figure(figsize=(6, 5))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc_val:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'Receiver Operating Characteristic (ROC) Curve for {name}')
    plt.legend(loc="lower right")
    plt.show()
```

**Reasoning**:
The `UndefinedMetricWarning` for Logistic Regression indicates that the model is not predicting any positive class instances, likely due to class imbalance. I will add `class_weight='balanced'` to the `LogisticRegression` model to mitigate this issue, which will automatically adjust weights inversely proportional to class frequencies, encouraging the model to consider the minority class.



```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix, roc_curve, auc
import matplotlib.pyplot as plt
import seaborn as sns

# 2. Create a dictionary to store the trained models and another dictionary to store their performance metrics.
models = {}
metrics = {}

# 3. Instantiate each model with random_state=42 for reproducibility.
rf_model = RandomForestClassifier(random_state=42)
dt_model = DecisionTreeClassifier(random_state=42)
lr_model = LogisticRegression(random_state=42, solver='liblinear', class_weight='balanced') # FIX: Added class_weight='balanced' to handle class imbalance

model_list = {
    "Random Forest": rf_model,
    "Decision Tree": dt_model,
    "Logistic Regression": lr_model
}

for name, model in model_list.items():
    print(f"\n--- Training and Evaluating {name} ---")

    # a. Train the model
    model.fit(X_train, y_train)

    # b. Make predictions on X_test
    y_pred = model.predict(X_test)
    y_pred_proba = model.predict_proba(X_test)[:, 1] # Probability of the positive class

    # c. Calculate and print evaluation metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, zero_division=0) # FIX: Added zero_division=0 to handle cases where precision is undefined
    recall = recall_score(y_test, y_pred, zero_division=0) # FIX: Added zero_division=0
    f1 = f1_score(y_test, y_pred, zero_division=0) # FIX: Added zero_division=0
    roc_auc = roc_auc_score(y_test, y_pred_proba)

    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1:.4f}")
    print(f"ROC AUC Score: {roc_auc:.4f}")

    # d. Store these metrics in the metrics dictionary.
    metrics[name] = {
        'Accuracy': accuracy,
        'Precision': precision,
        'Recall': recall,
        'F1 Score': f1,
        'ROC AUC': roc_auc
    }
    models[name] = model

    # e. Generate and display a confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(6, 5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=['No Churn', 'Churn'], yticklabels=['No Churn', 'Churn'])
    plt.title(f'Confusion Matrix for {name}')
    plt.xlabel('Predicted Label')
    plt.ylabel('True Label')
    plt.show()

    # f. Calculate FPR, TPR and plot ROC curve
    fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba)
    roc_auc_val = auc(fpr, tpr)

    plt.figure(figsize=(6, 5))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc_val:.2f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'Receiver Operating Characteristic (ROC) Curve for {name}')
    plt.legend(loc="lower right")
    plt.show()
```
